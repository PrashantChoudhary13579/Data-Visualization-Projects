# Gender

```python
ax = sns.countplot(x="Gender",data =df)
plt.title("Gender Distribution")
```

```python
ax = sns.countplot(x="Gender",data =df)
plt.title("Gender Distribution")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales = df.groupby(["Gender"],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
```

```python
sales
```

```python
sns.barplot(x="Gender",y="Amount",data=sales)
```

```python
# from the above data analysis we can say that female invest more money on shopping . That's why our main customer are female.
```

# State

```python
sales_state = df.groupby(['State'],as_index=False).sum().sort_values(by="Orders",ascending=False)
sns.set(rc ={'figure.figsize':(16,5)})
sns.barplot(data = sales_state,x="State",y="Orders")
```

```python
sales_state = df.groupby(['State'],as_index=False).sum().sort_values(by="Orders",ascending=False).head(5)
# sns.set(rc ={'figure.figsize':(10,5)})
sns.barplot(data = sales_state,x="State",y="Orders",hue ="Orders",palette ='viridis')
```

```python
sales_state_amt = df.groupby(['State'],as_index=False).sum().sort_values(by="Amount",ascending=False).head(9)
sns.set(rc={'figure.figsize':(14,5)})
sns.barplot(data =sales_state_amt,x='State',y='Amount',hue="Amount")
```

```python
# From the above graphs of state we can analyse that Big states like uttarpradesh, maharastra, karnataka are having the highest orders as well as the highest purchasing power
```

# Occupation

```python
df.columns
```

```python
axis = sns.countplot(data = df, y="Occupation")
# sns.set(rc ={'figure:figsize':(15,5)})

for bars in axis.containers:
    axis.bar_label(bars)
```

```python
sns.set(rc ={'figure.figsize':(20,5)})
axis = sns.countplot(data = df, x="Occupation")

for bars in axis.containers:
    axis.bar_label(bars)
```

```python
occupation_count = df["Occupation"].value_counts().sort_values(ascending=False).head(5)
sorted_order = occupation_count.index
sns.set(rc ={'figure.figsize':(7,5)})
axis = sns.countplot(data = df, x ="Occupation",order = sorted_order)

for bars in axis.containers:
    axis.bar_label(bars)
plt.show()
```

```python
sales_occ = df.groupby(["Occupation"],as_index =False).sum().sort_values(by="Amount",ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x="Occupation", y="Amount",data = sales_occ)
```

```python
# This shows that the occupation which has maximum number of count has given the maximum amount of money for purchasing the product in different occupation.
```

# Product category

```python
sns.set(rc={"figure.figsize":(20,5)})

category_Count = df["Product_Category"].value_counts().sort_values(ascending=False)
sorted_count = category_Count.index

ax = sns.countplot(x="Product_Category",data = df,order=sorted_count)

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_category = df.groupby(["Product_Category"], as_index = False)["Amount"].sum().sort_values(by="Amount",ascending=False)

sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(data = sales_category,x ="Product_Category",y="Amount",hue="Product_Category")
```
